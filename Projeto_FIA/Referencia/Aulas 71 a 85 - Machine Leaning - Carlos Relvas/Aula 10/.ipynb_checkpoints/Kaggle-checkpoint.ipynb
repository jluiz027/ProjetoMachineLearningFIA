```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("fraud_train_label.csv", sep=",")
tests = pd.read_csv("ex.csv")
```

```python
df
```

```python
print(df.shape)
print(tests.shape)
```

```python
features = df.columns[2:]
```

```python
from sklearn.model_selection import train_test_split
X = df[features]
y = df.SeriousDlqin2yrs
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
missings = []
for i in features:
    num_miss = sum(X_train[i].isnull())
    if num_miss > 0:
        missings.append(i)
    print(i+' : '+str(num_miss))
```

```python
for j in missings:
    mediana = np.nanmedian(X_train[j])
    X_train.loc[X_train[j].isnull(),j] = mediana
    X_test.loc[X_test[j].isnull(),j] = mediana
    tests.loc[tests[j].isnull(),j] = mediana
```

```python
missings = []
for i in features:
    num_miss = sum(X_train[i].isnull())
    if num_miss > 0:
        missings.append(i)
    print(i+' : '+str(num_miss))
```

```python
features
```

```python
X_train['income_dep'] = X_train.MonthlyIncome/(X_train.NumberOfDependents+1)
X_train['income_age'] = X_train.MonthlyIncome/(X_train.age+1)
X_train['income_finan'] = X_train.MonthlyIncome/(X_train.DebtRatio+1)
X_train['60_90'] = X_train['NumberOfTime60.89DaysPastDueNotWorse']/(X_train['NumberOfTimes90DaysLate']+1)
X_train['30_90'] = X_train['NumberOfTime30.59DaysPastDueNotWorse']/(X_train['NumberOfTimes90DaysLate']+1)
X_train['30_60'] = X_train['NumberOfTime30.59DaysPastDueNotWorse']/(X_train['NumberOfTime60.89DaysPastDueNotWorse']+1)


X_test['income_dep'] = X_test.MonthlyIncome/(X_test.NumberOfDependents+1)
X_test['income_age'] = X_test.MonthlyIncome/(X_test.age+1)
X_test['income_finan'] = X_test.MonthlyIncome/(X_test.DebtRatio+1)
X_test['60_90'] = X_test['NumberOfTime60.89DaysPastDueNotWorse']/(X_test['NumberOfTimes90DaysLate']+1)
X_test['30_90'] = X_test['NumberOfTime30.59DaysPastDueNotWorse']/(X_test['NumberOfTimes90DaysLate']+1)
X_test['30_60'] = X_test['NumberOfTime30.59DaysPastDueNotWorse']/(X_test['NumberOfTime60.89DaysPastDueNotWorse']+1)


tests['income_dep'] = tests.MonthlyIncome/(tests.NumberOfDependents+1)
tests['income_age'] = tests.MonthlyIncome/(tests.age+1)
tests['income_finan'] = tests.MonthlyIncome/(tests.DebtRatio+1)
tests['60_90'] = tests['NumberOfTime60.89DaysPastDueNotWorse']/(tests['NumberOfTimes90DaysLate']+1)
tests['30_90'] = tests['NumberOfTime30.59DaysPastDueNotWorse']/(tests['NumberOfTimes90DaysLate']+1)
tests['30_60'] = tests['NumberOfTime30.59DaysPastDueNotWorse']/(tests['NumberOfTime60.89DaysPastDueNotWorse']+1)
```

```python
features = X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
to_tests = scaler.transform(tests[features])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
```

```python
tuned_parameters = [{'hidden_layer_sizes': [(10,10),(20,20),(30,30)],
                     'activation' : ['logistic'],
                     'learning_rate': ['adaptive'],
                     'alpha': [0.001, 0.01, 0.02], 
                     'learning_rate_init': [0.001]}]
```

```python
nn = GridSearchCV(MLPClassifier(random_state=42,), tuned_parameters, cv=4)
nn.fit(X_train, y_train)
```

```python
print(nn.best_params_)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
tuned_parameters = [{'n_estimators': [100],
                     'max_depth' : [3, 5],
                     'min_samples_split': [10, 30],
                     'learning_rate': [0.01, 0.1, 1], 
                     'subsample': [1]}]
```

```python
bos = GridSearchCV(GradientBoostingClassifier(random_state=42), tuned_parameters, cv=5)
bos.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
tuned_parameters = [{'n_estimators': [100,300],
                     'max_depth': [2,5,10],
                     'max_features': [2,5,10,15]}]

rf = GridSearchCV(RandomForestClassifier(), tuned_parameters, cv=5)
rf.fit(X_train, y_train)
```

```python
print(bos.best_params_)
print(rf.best_params_)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
print(roc_auc_score(y_test, nn.predict_proba(X_test)[:,1]))
print(roc_auc_score(y_test, bos.predict_proba(X_test)[:,1]))
print(roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))
print(roc_auc_score(y_test, (nn.predict_proba(X_test)[:,1]+
                             bos.predict_proba(X_test)[:,1]+
                             rf.predict_proba(X_test)[:,1])/3))
```

```python
print(roc_auc_score(tests.SeriousDlqin2yrs, nn.predict_proba(to_tests)[:,1]))
print(roc_auc_score(tests.SeriousDlqin2yrs, bos.predict_proba(to_tests)[:,1]))
print(roc_auc_score(tests.SeriousDlqin2yrs, rf.predict_proba(to_tests)[:,1]))
print(roc_auc_score(tests.SeriousDlqin2yrs, (
                             bos.predict_proba(to_tests)[:,1]+
                             rf.predict_proba(to_tests)[:,1])/2))
```
